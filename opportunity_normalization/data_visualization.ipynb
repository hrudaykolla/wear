```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {'x_values': [1, 2, 3, 4, 5, 2, 3, 4, 5, 6],
        'label': ['A', 'A', 'B', 'A', 'B', 'B', 'A', 'B', 'A', 'A']}
df = pd.DataFrame(data)

# Create box plots
plt.figure(figsize=(8, 6))
df.boxplot(column='x_values', by='label', grid=False)

# Set plot labels and title
plt.xlabel('Label')
plt.ylabel('X Values')
plt.title('Box Plot of X Values Grouped by Label')

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame
data = {'x_values': [1, 2, 3, 4, 5, 2, 3, 4, 5, 6],
        'label': ['A', 'A', 'B', 'A', 'B', 'B', 'A', 'B', 'A', 'A']}
df = pd.DataFrame(data)

# Create horizontal box plots
plt.figure(figsize=(8, 6))
df.boxplot(column='x_values', by='label', grid=False, vert=False)

# Set plot labels and title
plt.xlabel('X Values')
plt.ylabel('Label')
plt.title('Horizontal Box Plot of X Values Grouped by Label')

# Show the plot
plt.show()
```
